# Opportunity Zone Data Clean-Up

```python
# Imports
import tabula
import pandas as pd
from pathlib import Path
```

```python
# Read pdf file
df_oz_raw = tabula.read_pdf("Data/opp_zone.pdf",pandas_options={'header': None},pages='all')
df_oz_raw.rename(columns={0:'census_tract_nbr',1:'map_nbr',2:'county',3:'neighborhood'},inplace=True)
df_oz_raw
```

```python
df_oz_junction = pd.read_csv(Path('JUNCTION_TABLE/junction_oppzone_neighborhoods.csv'))
df_oz_junction
```

```python
df_oz = pd.merge(df_oz_junction,df_oz_raw,left_on='Neighborhood_op',right_on='neighborhood',how='left')
df_oz.loc[df_oz['census_tract_nbr'].isnull(),'op_zone'] = 0
df_oz.loc[df_oz['census_tract_nbr'].notnull(),'op_zone'] = 1
df_oz.drop(columns=(['Neighborhood_op','census_tract_nbr','map_nbr','county','neighborhood']),inplace=True)
df_oz.drop_duplicates(subset = 'Neighborhood', keep = 'first', inplace=True)
df_oz.set_index(['Neighborhood'],inplace=True)
```

```python
df_oz.to_csv('joint_analysis_resources/opp_zone_data.csv')
```
